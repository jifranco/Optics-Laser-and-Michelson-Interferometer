```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the Gaussian function
def gaus(delta_x, w):
    return np.exp(-2.*delta_x**2/w**2)

# Define the model function
def model_func(x, xmean, w, I0):
    I = np.zeros(len(x))
    for i in range(0, len(x)):
        delta_x = x[i] - xmean
        I[i] = I0*quad(gaus, delta_x, np.inf, args=(w,))[0]
    return I

# X-values
x_values = np.array(inches_list_10cm) 

# Initial guess for the parameters
initial_guess = [0, 1, max(power_data_10cm)]  # Example initial guess including I0

# Perform the curve fitting
params, covariance = curve_fit(model_func, x_values, power_data_10cm, p0=initial_guess)

# Extract the fitted w value for z = 10cm
w_10cm = params[1]

# Generate fitted curve data
fitted_curve = model_func(x_values, *params)

# Plot the data and the fitted curve
plt.figure(figsize=(10, 6))
plt.plot(x_values, power_data_10cm, 'bo', label='Data')
plt.plot(x_values, fitted_curve, 'r-', label='Fitted Curve')
plt.title('Power Measurement Fitting for z=10cm')
plt.xlabel('Distance (inches)')
plt.ylabel('Power (mW)')
plt.legend()
plt.show()

print(w_10cm)
```
